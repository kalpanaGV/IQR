```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("Placement.csv")
```

```python
dataset
```

```python
from uni import univariate
obj=univariate()
quan,qual=obj.QuanQual(dataset)
```

```python
dataset["salary"]=dataset["salary"].fillna(0)
```

```python
univ=obj.uniAnalysis(dataset,quan)
```

```python
univ
```

```python
univ["ssc_p"]["Lesser"]<univ["ssc_p"]["Min"]
```

```python

lesser=[]
greater=[]

for columName in quan:
    if(univ[columName]["Lesser"]>univ[columName]["Min"]):
        lesser.append(columName)
    if(univ[columName]["Greater"]<univ[columName]["Max"]):
        greater.append(columName)
        
```

```python
lesser
```

```python
greater
```

```python
dataset
```

```python
for lesscolumn in lesser:
    dataset[lesscolumn][dataset[lesscolumn]<univ[lesscolumn]["Lesser"]]=univ[lesscolumn]["Lesser"]
for greatercolum in greater:
    dataset[greatercolum][dataset[greatercolum]>univ[greatercolum]["Greater"]]=univ[greatercolum]["Greater"]
```

```python
univ_after=obj.uniAnalysis(dataset,quan)
```

```python
univ_after
```

```python
lesser=[]
greater=[]

for columName in quan:
    if(univ_after[columName]["Lesser"]>univ_after[columName]["Min"]):
        greater.append(columName)
    if(univ_after[columName]["Greater"]<univ_after[columName]["Max"]):
        greater.append(columName)
```
